df[col].mode() returns a series of the most frequent value(s) in the column, and [0] selects the first mode (in case there are multiple modes, though in most cases there will be only one).

```python
def remove_outliers(column):
    Q1 = column.quantile(0.25)  # First quartile (25th percentile)
    Q3 = column.quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1               # Interquartile range
    threshold = 1.5 * IQR       # Define the threshold for outliers (1.5 times the IQR)
    outlier_mask = (column < Q1 - threshold) | (column > Q3 + threshold)  # Identify outliers
    return column[~outlier_mask]  # Return the column with outliers removed
```

The error TypeError: unsupported operand type(s) for -: 'str' and 'str' occurs because the function remove_outliers() is attempting to perform a mathematical operation (such as subtracting or comparing) on string data (which is not supported for numerical calculations).

```python
for col in df.columns:
    # Check if the column is numeric (int or float)
    if df[col].dtype in ['float64', 'float32']:
        df[col] = remove_outliers(df[col])
```

MinMaxScaler is used for continuous numerical columns to normalize them into a range between 0 and 1.

LabelEncoder would not work here because it's intended for categorical features, not numerical ones.

Why use MinMaxScaler here?
so2, no2, rspm are numerical data (continuous values).

You scale them between 0 and 1 to ensure that your machine learning model doesn't give more importance to features with larger values (e.g., rspm could have a larger range than so2), which could negatively affect model performance.

Key columns that are already numerical are normalized for better model convergence.

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
encoder = LabelEncoder()
scaler = MinMaxScaler()
```

```python
categorical_cols = [ 'CO(GT)', 'C6H6(GT)', 'T', 'RH', 'AH']
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])

    numerical_cols = ['PT08.S1(CO)', 'NMHC(GT)', 'PT08.S2(NMHC)', 'NOx(GT)',
                  'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)']

# Apply MinMax Scaling for numerical columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Check the transformed data
print(df.head())
```

MinMax Scaling is useful when you need to normalize the range of numerical features (e.g., transforming all values to be between 0 and 1). This is important when working with models that are sensitive to the scale of features, such as neural networks, SVM, and K-means clustering.

Label Encoding is needed when you have categorical columns (non-numeric columns) and you want to use them as input for machine learning models. Many machine learning algorithms, such as Logistic Regression, SVM, or Random Forest, require numerical input, so we need to convert categorical data (like strings) into numbers.

```python
# **Linear Regression Model**
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)  # Model ko train karte hain
lr_pred = lr_model.predict(X_test)  # Test data par predictions karte hain

# **Decision Tree Regressor Model**
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)  # Model ko train karte hain
dt_pred = dt_model.predict(X_test)  # Test data par predictions karte hain

# Model ki performance evaluate karte hain
def evaluate_model(name, y_true, y_pred):
    r2 = r2_score(y_true, y_pred)  # R-squared score
    mae = mean_absolute_error(y_true, y_pred)  # Mean absolute error
    mse = mean_squared_error(y_true, y_pred)  # Mean squared error
    print(f"\n{name} Model:")
    print(f" RÂ² Score     : {r2:.4f}")
    print(f" MAE          : {mae:.4f}")
    print(f" MSE          : {mse:.4f}")
    print(f" Accuracy (%) : {r2*100:.2f}%")

# Models ki evaluation karte hain
evaluate_model("Linear Regression", y_test, lr_pred)
evaluate_model("Decision Tree Regressor", y_test, dt_pred)
```

r2_score(y_test, y_pred) gives the RÂ² value, which indicates how well the model fits the data.

By multiplying r2 by 100, we convert it into percentage form.

The .2f formatting ensures that the result is printed with two decimal places.
